```python
# 70대 남성

pb = [1.602, 2.041, 2.452, 3.951]
hg = [1.965, 2.939, 5.189, 9.648]
cd = [.703, 0.966, 1.245, 2.056]

data= [pb, hg, cd]

user = [2.443, 1.367, 1.574]

def metal_point (data, user): 
    result = [0, 0, 0]
    
    for i in range(3):
        d = data[i]
        u = user[i]
        
        if u < d[0]:
            result[i] = u / d[0] * 25
        elif u < d[1]:
            result[i] = (u - d[0]) / (d[1] - d[0]) * 25 + 25
        elif u < d[2]:
            result[i] = (u - d[1]) / (d[2] - d[1]) * 25 + 50
        elif u < d[3] :
            result[i] = (u - d[2]) / (d[3] - d[2]) * 20 + 75
        else :
            result[i] = 99
    
    return result
```

```python
metal_point(data, user)
```
